Step 2: Feature Extraction

```python
# Check the column names in the dataset
print(df.columns)
```

```python
df['genre'] = df['genre'].apply(lambda x: x.split(','))
```

```python
genre_encoded = df['genre'].str.join('|').str.get_dummies()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

3. Recommendation System:

Cosine Similarity Calculation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
features = pd.concat([df[['rating', 'members']], genre_encoded], axis=1)
```

```python
cosine_sim = cosine_similarity(features)
```

```python
def recommend_anime(anime_id, cosine_sim,df):
    idx = df[df['anime_id'] == anime_title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    anime_indices = [i[0] for i in sim_scores[1:11]]  # Get top 10 similar anime
    return df['anime_id'].iloc[anime_indices]
```

4. Evaluation:

```python
#train-test-split
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

Evaluate the Recommendation System

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
def evaluate_recommendations(true_relevant, recommended):
    precision = precision_score(true_relevant, recommended)
    recall = recall_score(true_relevant, recommended)
    f1 = f1_score(true_relevant, recommended)

    return precision, recall, f1
```

```python
import numpy as np
true_relevant = np.array([1, 0, 1, 1, 0, 0, 1, 0, 1, 0])
recommended = np.array([1, 1, 0, 1, 0, 0, 1, 0, 0, 0])
```

```python
precision, recall, f1 = evaluate_recommendations(true_relevant, recommended)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```

#Explanation of Code:
Data Preprocessing: We load the dataset and handle missing values. For categorical columns like genre, we perform one-hot encoding to convert them into numerical format. Numerical features like average_rating and num_episodes are normalized to have a similar scale.

Cosine Similarity: Using cosine_similarity(), we compute similarity scores between each anime based on the selected features. The recommend_anime function retrieves the top N similar anime based on these scores.

Evaluation: The evaluation part is simplified by splitting the dataset into training and testing sets and calculating precision, recall, and F1-score.
